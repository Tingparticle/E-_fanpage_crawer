ä»¥likesæ•¸ç›®ä¾†çœ‹ç†±é–€ã€å†·é–€è©å½™

```python
data_30_up = data.loc[data.likes > 30, :]

for i in data_30_up.content.index:
    hash = {}
    for item in jieba.cut(data_30_up.content[i], cut_all=False): 
        if item in hash:
            hash[item] += 1
        else:
            hash[item] = 1

    fd_30_up = open("count_30_up.csv","w")
    fd_30_up.write("word,count\n")
    for k in hash:
        fd_30_up.write("%s,%d\n"%(k, hash[k]))

data_10_down = data.loc[data.likes > 10, :]

for i in data_10_down.content.index:
    if str(data_10_down.content[i]) == str('nan'):
        del data_10_down.content[i]

for i in data_10_down.content.index:
    hash = {}
    for item in jieba.cut(data_10_down.content[i], cut_all=False): 
        if item in hash:
            hash[item] += 1
        else:
            hash[item] = 1

    fd_10_down = open("count_10_down.csv","w")
    fd_10_down.write("word,count\n")
    for k in hash:
        fd_10_down.write("%s,%d\n"%(k, hash[k]))
#fd.write("%s,%d\n"%(k.encode("utf8"),hash[k]))
```

```python
count_30_up = pd.read_csv('count_30_up.csv')

count_10_down = pd.read_csv('count_10_down.csv', sep='delimiter')

cou = []
for i in range(0, pd.read_csv('count_10_down.csv', sep='delimiter').shape[0]):
    cou.append(int(re.search(r'\d+' ,str(pd.read_csv('count_10_down.csv', sep='delimiter').iloc[i, :])).group()))
cou = pd.DataFrame(cou)
count_10_down = pd.concat([count_10_down, cou] , axis=1)
count_10_down.to_csv('count_10_down.csv')
count_10_down = pd.read_csv('count_10_down.csv')
```
